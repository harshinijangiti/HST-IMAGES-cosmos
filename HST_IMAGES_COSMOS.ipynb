### 1)F125 file

```python
#!/usr/bin/python3
#this program creates a cutout of size 20 arcsec X 20 arcsec around RA,DEC=53.09964, -27.79427 from the F125 image of goods -south data
# this script works with gzipped fits file also
from astropy.nddata import Cutout2D
from astropy import units as u
zoomSize = u.Quantity((20,20), u.arcsec)
from astropy import wcs
from astropy.io import fits
f1name=r'/home/harshini/Downloads/cosmos_3dhst.v4.0.F125W_orig_sci.fits'
hdulist1 = fits.open(f1name)
w1 = wcs.WCS(hdulist1[0].header)
data1=fits.getdata(f1name,ext=0)
from astropy.coordinates import SkyCoord
position1 = SkyCoord(150.10114,2.1906323, unit="deg")
cutout1 = Cutout2D(data1, position1, zoomSize, wcs=w1)
xdu1 = fits.PrimaryHDU(cutout1.data)
xdu1.header = hdulist1[0].header
xdu1.header.update(cutout1.wcs.to_header())
xdu1.writeto(r'/home/harshini/Downloads/PROJECT/output_files/cosmos_3dhst.v4.0.F125W_orig_sci_COS4_00779.fits',overwrite=True)
```
